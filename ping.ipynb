```python
import pygame
import math
import random
import time

pygame.init()
screen=pygame.display.set_mode((800,600))
clock = pygame.time.Clock()

pygame.display.set_caption("Ping")
#attribution of icon Icons made by <a href="https://www.flaticon.com/authors/freepik" title="Freepik">Freepik</a> from <a href="https://www.flaticon.com/" title="Flaticon"> www.flaticon.com</a>
time.sleep(5)

def ball(x,y,surface):
    pygame.draw.circle(surface,(255,0,0),(x,y),10,10)
#     surface.blit(ball_img,(x-16,y-16)) #icon made by freepik at flaticon.com
    
def player(start,end,rect,surface):
    pygame.draw.arc(surface,(0, 225, 218),rect,start,end,11)

#define rectangle for arcs
rect=pygame.draw.rect(screen, (0,0,0), (200,100,400,400),1)
start_arc=math.pi/2-.3
end_arc=math.pi/2+.3
increment=0

#score
score=0
font=pygame.font.Font(pygame.font.get_default_font(),32)
scorex=760
scorey=20
def showscore(score):
    scoreboard=font.render('Score: ' + str(score), True, (255,255,255))
    screen.blit(scoreboard, (600,20))
def gameover():
    purple=pygame.font.Font(pygame.font.get_default_font(),32)
    game_over=purple.render('Game over', True, (255,255,255))
    screen.blit(game_over,(400,300))
    
    


#ball initilization
ballx=400
bally=300
# ballv_x=random.randint(-2,2)
# ballv_y=random.randint(-2,2)
ballv_x=1
ballv_y=0
if ballv_x ==0 and ballv_y==0:
    ballv_x==1

angle=0
def angle(x,y):
    if x==400 and y<300:
        return math.pi/2
    if x== 400 and y>300:
        return 3*math.pi/2
    if (x==400 and y==300):
        return 0
    ang=math.atan(abs(300-y)/abs(x-400))
    if (x-400)<0 and (300-y)>0:
#         print(ang)
        return math.pi-ang
    if ang==0 and (x-400)<0:
        return math.pi
    if x-400<0 and 300-y<0:
        return ang+math.pi
    if x-400>0 and 300-y<0:
        return 2*math.pi-ang
    else:
        return ang

        
running=True
while running:
    clock.tick(60)
    for event in pygame.event.get():
        if event.type==pygame.QUIT:
            running =False
            pygame.display.quit()
            pygame.quit()
        if event.type==pygame.KEYDOWN:
            if event.key==pygame.K_LEFT:
                increment=.1
            elif event.key==pygame.K_RIGHT:
                increment=-.1
        if event.type==pygame.KEYUP:
            if event.key==pygame.K_LEFT or event.key==pygame.K_RIGHT:
                 increment=0
            
    if running:         
        start_arc+=increment
        if start_arc>2*math.pi:
            start_arc-=2*math.pi
        end_arc+=increment
        if end_arc>2*math.pi:
            end_arc-=2*math.pi
        if start_arc<0:
            start_arc+=2*math.pi
        if end_arc<0:
            end_arc+=2*math.pi        
        
        ballx+=ballv_x
        bally+=ballv_y
        
        ang=angle(ballx,bally)
#         print(start_arc)
#         print(end_arc)
#         print(ang)
    #horizontals are not working properly
    #
    
        if 220>= math.sqrt((ballx-400)**2+(300-bally)**2)>=200:
            if start_arc<ang<end_arc:
                ballv_x=-ballv_x
                ballv_y=-ballv_y
                score+=1
            if end_arc<start_arc and bally==300:
                ballv_x=-ballv_x
                ballv_y=-ballv_y
                score+=1
                
        if ballx<=0 or ballx>=800 or bally<=0 or bally>=600:
            running=False
            pygame.display.quit()
            pygame.quit()
            time.sleep(3)
            break
            
            #state goal is to beat the ball there if they are moving there is a chance that they will miss
            
        
                

        
            
       
                
        if bally == 300 and ballx==400:
            ballv_x+=random.randint(-2,2)
            ballv_y+=random.randint(-2,2)
            if ballv_x==0 and ballv_y==0:
                ballv_x=1
        screen.fill((255, 167, 129))
        pygame.draw.circle(screen,(91, 14, 45),(400,300),200,10)
        player(start_arc,end_arc,rect,screen)
        ball(ballx,bally,screen)
        showscore(score)
        pygame.display.update()
```
